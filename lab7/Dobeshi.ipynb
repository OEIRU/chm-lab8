```python
import numpy as np
import matplotlib.pyplot as plt

#    
data = np.loadtxt('tests/Dobeshi_Z.txt')

#        
plt.figure(figsize=(10, 6))

#        
plt.plot(data, linestyle='-', label='')  #  marker='o'

#   Y      
plt.ylim(0.999, 1.001)  #    Y

#   
plt.xlabel('')
plt.ylabel('')
plt.title('  -   ')
plt.legend()
plt.grid(True)

#  
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(data, linestyle='-', label='')  #  marker='o'
plt.yscale('log')  #     Y
plt.xlabel('')
plt.ylabel(' ( )')
plt.title('  -  ')
plt.legend()
plt.grid(True)
plt.show()
```
