```python
import numpy as np
import matplotlib.pyplot as plt

# Загрузка данных из файла
data = np.loadtxt('tests/Dobeshi_Z.txt')

# Построение графика с акцентом на значения около единицы
plt.figure(figsize=(10, 6))

# Основной график с акцентом на значения около единицы
plt.plot(data, linestyle='-', label='Данные')  # Убрали marker='o'

# Ограничение оси Y для акцента на значениях около единицы
plt.ylim(0.999, 1.001)  # Устанавливаем диапазон оси Y

# Подписи и легенда
plt.xlabel('Индекс')
plt.ylabel('Значение')
plt.title('Исходный сигнал - значения около единицы')
plt.legend()
plt.grid(True)

# Показать график
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(data, linestyle='-', label='Данные')  # Убрали marker='o'
plt.yscale('log')  # Логарифмическая шкала по оси Y
plt.xlabel('Индекс')
plt.ylabel('Значение (логарифмическая шкала)')
plt.title('Исходный сигнал - логарифмический график')
plt.legend()
plt.grid(True)
plt.show()
```
