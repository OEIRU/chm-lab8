```python
import pandas as pd
import matplotlib.pyplot as plt

#    CSV 
prey_data = pd.read_csv('prey_population.csv')
predator_data = pd.read_csv('predator_population.csv')
phase_data = pd.read_csv('phase_portrait.csv')

#  
plt.figure(figsize=(12, 8))

#  1:   
plt.subplot(2, 2, 1)
plt.plot(prey_data['Day'], prey_data['Prey'], color='blue', label='')
plt.title('  ')
plt.xlabel('')
plt.ylabel(' ')
plt.grid()
plt.legend()

#  2:   
plt.subplot(2, 2, 2)
plt.plot(predator_data['Day'], predator_data['Predator'], color='red', label='')
plt.title('  ')
plt.xlabel('')
plt.ylabel(' ')
plt.grid()
plt.legend()

#  3:   
plt.subplot(2, 1, 2)
plt.plot(phase_data['Prey'], phase_data['Predator'], color='green', label=' ')
plt.title('   ')
plt.xlabel(' ')
plt.ylabel(' ')
plt.grid()
plt.legend()

#  
plt.tight_layout()
plt.show()
```
