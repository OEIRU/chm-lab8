```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из CSV файлов
prey_data = pd.read_csv('prey_population.csv')
predator_data = pd.read_csv('predator_population.csv')
phase_data = pd.read_csv('phase_portrait.csv')

# Настройка графиков
plt.figure(figsize=(12, 8))

# График 1: Динамика численности жертв
plt.subplot(2, 2, 1)
plt.plot(prey_data['Day'], prey_data['Prey'], color='blue', label='Жертвы')
plt.title('Динамика численности жертв')
plt.xlabel('День')
plt.ylabel('Численность жертв')
plt.grid()
plt.legend()

# График 2: Динамика численности хищников
plt.subplot(2, 2, 2)
plt.plot(predator_data['Day'], predator_data['Predator'], color='red', label='Хищники')
plt.title('Динамика численности хищников')
plt.xlabel('День')
plt.ylabel('Численность хищников')
plt.grid()
plt.legend()

# График 3: Фазовый портрет взаимодействия
plt.subplot(2, 1, 2)
plt.plot(phase_data['Prey'], phase_data['Predator'], color='green', label='Фазовый портрет')
plt.title('Фазовый портрет взаимодействия популяций')
plt.xlabel('Численность жертв')
plt.ylabel('Численность хищников')
plt.grid()
plt.legend()

# Отображение графиков
plt.tight_layout()
plt.show()
```
